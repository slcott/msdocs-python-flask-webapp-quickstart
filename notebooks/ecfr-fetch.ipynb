```python
base = "https://www.ecfr.gov/api/versioner/v1"
# url = base + "/current/title-7/subtitle-B/chapter-II/subchapter-A/part-210"
# url = base + "/structure/2024-12-11/title-7"
url = base + "/versions/title-7"
response = requests.get(url)
print(response)
```

```python
print(response.text[:500])
```

```python
d = response.json()
d
```

```python
versions = d['content_versions'][:10]
[v for v in versions if v['title'] == '7']
```
